```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

```

```python
from langchain_core.documents import Document

# Create LangChain documents for IPL players

doc1 = Document(
        page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and fitness, he has led the Royal Challengers Bangalore in multiple seasons.",
        metadata={"team": "Royal Challengers Bangalore"}
    )
doc2 = Document(
        page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm demeanor and ability to play big innings under pressure.",
        metadata={"team": "Mumbai Indians"}
    )
doc3 = Document(
        page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and leadership are legendary.",
        metadata={"team": "Chennai Super Kings"}
    )
doc4 = Document(
        page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death-over expertise.",
        metadata={"team": "Mumbai Indians"}
    )
doc5 = Document(
        page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding and match-winning performances make him a key player.",
        metadata={"team": "Chennai Super Kings"}
    )
```

```python
# creating our hugging face embedding model
Embedding_model = HuggingFaceEmbeddings(model ="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# creating a list to keep all these docs
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
# creation of vector store
vector_store = Chroma(
    embedding_function = Embedding_model,
    persist_directory='my_chroma_db',
    collection_name='sample'
)
```

```python
# adding into vector store
vector_store.add_documents(docs)
```

```python
# now lets view the documents
# yu will pass few parameters in the .get function
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
# searching in the vector store
vector_store.similarity_search(
    query='who among these are a bowler',
    k=2 #kitne objects deikhana chatte ho
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query='who among these are a bowler',
    k=2 #kitne objects deikhana chatte ho
)  #score jitna kam utna aacha
```

```python
#meta-data filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Chennai Super Kings"}
)
```

```python
# update documents
update_doc1 = Document(
        page_content="VK most successful and consistent batsmen in IPL history. Known for his aggressive batting style and fitness, he has led the Royal Challengers Bangalore in multiple seasons.",
        metadata={"team": "Royal Challengers Bangalore"}
    )
vector_store.update_document(document_id='47db042a-9abb-4bf9-865d-fc9c4a74c641', document=update_doc1)
```

```python
# viewing the updated docuemnt
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
# deleting the document
vector_store.delete(ids=['47db042a-9abb-4bf9-865d-fc9c4a74c641'])
```
