# Importing from the URL using the link

```python
import requests
from io import StringIO

url = "paste yur url here"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.4; RV:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# Sep Parameter

# Sep Parameter by default this parameter is sep=',' but for the tsv you can use \t 
# and names=['','',''] is used when there is no column names

```python
pd.read_csv('link of the csv', sep='\t', names=['sno','name','release_year','rating','votes','genres'])
```

# Index_col parameter (suppose you dont want that column or start from somewhere)

```python
pd.read_csv('creditcard_2023.csv', index_col='id')
```

```python
pd.read_csv('creditcard_2023.csv', header=1) #if the columns are being treated as rows then use this 
```

# use_cols parameter (when you want to use selective columns)

```python
pd.read_csv('creditcard_2023.csv', usecols=['V1','V5'])
```

```python
pd.read_csv('creditcard_2023.csv', skiprows=[0,1])
```
