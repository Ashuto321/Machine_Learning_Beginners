```python
import pandas as pd
import numpy as np
```

```python
# loading the data
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
# lets separate the columns which are important to us for
# ordinal on the input dataset and label encoding on output dataset.
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
# we will separate the training and testing data 1.rule before doing featuring
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.2)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# creating the Object for the ordinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average', 'Good'],['School', 'UG', 'PG']])
```

# we fir on train and we transform on train and test both

```python
# now fitting the object on training data
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
# we can also use categories feature like
oe.categories_
```

# Now we will work with the purchased column that is categorical in output column we will use label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# creating object for it
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
# To check which one is getting what
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
